```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Carregar e processar
img = np.array(Image.open('/content/drive/MyDrive/Lenna.png').convert('L'))
sobel_h = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

# Aplicar filtro
bordas = np.array([np.sum(img[i-1:i+2, j-1:j+2] * sobel_h)
                  for i in range(1, img.shape[0]-1)
                  for j in range(1, img.shape[1]-1)])
bordas = bordas.reshape(img.shape[0]-2, img.shape[1]-2)
bordas = (np.abs(bordas) / np.abs(bordas).max() * 255).astype(np.uint8)

# Mostrar resultados
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.imshow(Image.open('/content/drive/MyDrive/Lenna.png'))
ax1.set_title('Original', fontweight='bold')
ax1.axis('off')

ax2.imshow(bordas, cmap='gray')
ax2.set_title('Sobel Horizontal', fontweight='bold')
ax2.axis('off')

plt.tight_layout()
plt.show()
```
